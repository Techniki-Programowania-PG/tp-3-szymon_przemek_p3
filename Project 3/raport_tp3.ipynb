```python
#run this before everything else
import _core as mpl
from IPython.display import Image, display

r = 2*3.14 #constant - don't touch
sizeX = 560 #width of images
sizeY = 480 #height of images
```

Discrete fourier transform and its inverse

additional -> frequency function which shows individual frequencies of the original signals

```python
# frequencies: 1, 2, 4, 6, 8.5
a1 = mpl.sin(1,0,6*r,1000) * 3
a2 = mpl.sin(2,0,6*r,1000) * 1.5
a3 = mpl.cos(4,0,6*r,1000) * 0.5
a4 = mpl.sin(6,0,6*r,1000) * 2
a5 = mpl.cos(8.5,0,6*r,1000)

a = a1+a2+a3+a4+a5 #signal of sine/cosine functions added up

b = mpl.fourier(a) # discrete fourier transform
c = mpl.frequenciesDFT(b,10,1000,6) # rescale to show individual frequencies

d = mpl.inv_fourier(b) # inverse discrete fourier transform

mpl.plotIMG(a, "fourier1.png", "Signal", sizeX, sizeY)
mpl.plotIMG(b, "fourier2.png", "Fourier Transform", sizeX, sizeY)
mpl.plotIMG(c, "fourier3.png", "Frequencies Plot", sizeX, sizeY)
mpl.plotIMG(d, "fourier4.png", "Inverse Fourier Transform", sizeX, sizeY)
```

```python
display(Image("fourier1.png"))
display(Image("fourier2.png"))
display(Image("fourier3.png"))
display(Image("fourier4.png"))
```

1D filter

```python
e = mpl.sin(3,0,10,5000) + mpl.cos(4,0,10,5000) #1D signal
f = mpl.noise(e, 4) #apply gaussian noise

g = f
for i in range(50):
    g = mpl.filter1D(g) #1D filter 50x

mpl.plotIMG(e, "1DfilterOriginal.png", "1D Signal", sizeX, sizeY)
mpl.plotIMG(f, "1DfilterGauss.png", "Gaussian Noise", sizeX, sizeY)
mpl.plotIMG(g, "1DfilterFilter.png", "1D Filter", sizeX, sizeY)
```

```python
display(Image("1DfilterOriginal.png"))
display(Image("1DfilterGauss.png"))
display(Image("1DfilterFilter.png"))
```

2D filter

```python
a1 = 2 * mpl.sin(1,0,8,100)
a2 = -1 * mpl.sin(2,0,8,100)
a = mpl.multiplyPlots2D(a1,a2,0,8,0,8,100) #spacially multiply a1 with a2 to get a 2D signal

b = mpl.noise2D(a,4) #apply gaussian noise
c = b
for i in range(3):
    c = mpl.filter2D(c,100) #2D filter 3x

mpl.surfIMG(a,"2DsignalOriginal.png", "2D Signal",sizeX,sizeY)
mpl.surfIMG(b,"2DsignalGauss.png", "Gaussian Noise",sizeX,sizeY)
mpl.surfIMG(c,"2DsignalFilter.png", "2D Filter",sizeX,sizeY)
```

```python
display(Image("2DsignalOriginal.png"))
display(Image("2DsignalGauss.png"))
display(Image("2DsignalFilter.png"))
```

Correlation between two signals

```python
a = mpl.sin(3,0,16,5000)
b = mpl.sin(7,0,16,5000)

mpl.plotIMG(a,"corr1.png", "Signal 1", sizeX, sizeY)
mpl.plotIMG(b,"corr2.png", "Signal 2", sizeX, sizeY)

mpl.plotIMG(mpl.correlation(a,b),"corr3.png", "Correlation between Signal 1 and 2", sizeX, sizeY)
```
