DFT

```python
freq = 10
samples = 1000
sinus = mpl.sin(freq,0,12.56,samples)
dft_sin = mpl.fourier(sinus,0,12.56,samples)
mpl.plotIMG(dft_sin, "ftsin.png", "DFT of sinus", 560, 480)
```

```python
display(Image("ftsin.png"))
```
